```python
from joblib import load
import pandas as pd

# Load model parameters
params = load("model_parameters.joblib")
import matplotlib.pyplot as plt

coefs = params['coef'][0]  # First row of coefficients
features = ["Weight", "Smoking"]

plt.bar(features, coefs)
plt.title("Model Coefficients")
plt.ylabel("Value")
plt.grid(True)
plt.show()

print(f"Intercept: {params['intercept'][0]}")

colors = ['red' if c < 0 else 'green' for c in coefs]
plt.bar(features, coefs, color=colors)


```

## Let's test the model!

```python
from joblib import load

# Load model parameters
params = load("model_parameters.joblib")
coefs = params['coef'][0]
intercept = params['intercept'][0]

# Prompt user for input
weight = float(input("Enter patient's weight (e.g., 70): "))
smoking = int(input("Is the patient a smoker? (1 = Yes, 0 = No): "))

# Create input feature vector
x = [weight, smoking]

# Compute prediction
prediction = sum(w * xi for w, xi in zip(coefs, x)) + intercept
print(f"\nPredicted value: {prediction:.5f}")
```
